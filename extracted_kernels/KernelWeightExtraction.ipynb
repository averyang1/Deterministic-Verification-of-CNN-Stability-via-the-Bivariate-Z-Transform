```python
import torch
import torchvision.models as models
import numpy as np
import os
```

```python
def extractandSaveKernels(outputDir = "extractedKernels"):
    if not os.path.exists(outputDir):
        os.makedirs(outputDir)

#load in the pretrained ResNet-18 model and extract the kernels from the first convolutional layer of the first block in layer1
    model = models.resnet18(weights="IMAGENET1K_V1")


    #Target the first convolutional layer of the first block in layer1, which is a 3x3 kernel
    targetLayer = model.layer1[0].conv1.weight.detach()


    #shape of targetLayer is (64, 64, 3, 3) which corresponds to (out_channels, in_channels, kernel_height, kernel_width)
    kernelsNumpy = targetLayer.cpu().numpy()

    #to save as .npy
    npyPath = os.path.join(outputDir, "resnet18_layer1_3x3.npy")
    np.save(npyPath, kernelsNumpy)

    #to save csv
    csvPath = os.path.join(outputDir, "resnet18_layer1_3x3.csv")
    flattenedKernels = kernelsNumpy.reshape(-1,9)
    np.savetxt(csvPath, flattenedKernels, delimiter=",", header="h00,h01,h02,h10,h11,h12,h20,h21,h22")
    return kernelsNumpy


if __name__ == "__main__":
    kernels = extractandSaveKernels()
    
```
