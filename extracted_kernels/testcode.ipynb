```python
import torch 
import torchvision.models as models 
import numpy as np 
import matplotlib.pyplot as plt
import scipy.fft as fft
```

step 1: kernel extraction 
objective: extract learned weights from ResNet18

```python
def extractResnetKernels():
    #load pretrained resnet18 model
    model = models.resnet18(weights = "IMAGENET1K_V1")

    #extract first conv layer kernels
    kernels = model.layer1[0].conv1.weight.data.numpy()

    #select a sample kernel and normalize
    sampleKernel = kernels[0,0,:,:]
    sampleKernel = sampleKernel / np.sqrt(np.sum(sampleKernel**2))

    
    return sampleKernel
```

Step 2: 2D Z-Transform & Spectral Mapping
